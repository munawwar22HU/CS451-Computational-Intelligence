```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def Calculate(filename):
    df_bsf = pd.read_csv(filename+'_BSF.csv')
    df_bsf.drop(['Unnamed: 0', 'Avg BSF'],inplace=True, axis=1)
    array_bsf = df_bsf.to_numpy()
    new_array_bsf = np.zeros((array_bsf.shape))
    for i in range(1000):
        new_array_bsf[i] = np.max(array_bsf[:i+1,:10],axis=0)

    df_avg = pd.read_csv(filename+'_AVG.csv')
    df_avg.drop(['Unnamed: 0', 'Avg ASF'],inplace=True, axis=1)
    array_avg = df_avg.to_numpy()
    new_array_avg = np.zeros((array_avg.shape))
    for i in range(1000):
        new_array_avg[i] = np.mean(array_avg[:i+1,:10],axis=0)
    y1 = np.mean(new_array_bsf,axis=1)
    y2 = np.mean(new_array_avg,axis=1)
    return y1,y2
    
```

```python
def PlotGraph(filename,title):
    y1,y2 = Calculate(filename)
    x = np.arange(1000)

    plt.plot(x,y1)
    plt.xlabel ('Number of Generations')
    plt.ylabel ('Average Best Fitness so far')
    plt.title(str(title))
    plt.savefig(filename+'_bsf.png')
    plt.show()

    plt.plot(x,y2)
    plt.xlabel ('Number of Generations')
    plt.ylabel ('Average Average Fitness so far')
    plt.title(str(title))
    plt.savefig(filename+'_avg.png')
    plt.show()
```

```python
PlotGraph('fps_fps','FPS and FPS')
```
